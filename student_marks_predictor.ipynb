```python
# IMPORT  ESSENTIAL   LIBRARIES
import  numpy  as  np
import  pandas  as  pd
import  matplotlib.pyplot  as  plt
```

```python
# import  dataset
df=pd.read_csv("student_info.csv")
df.head()
```

```python
# data   visualization
plt.scatter(x=df.study_hours,y=df.student_marks)
plt.title("Student  Marks Predictor")
plt.xlabel("Study Hours")
plt.ylabel("Student Marks")
plt.show()
```

```python
#  check  missing  values
df.isnull().sum()
```

```python
df.info()
```

```python
#handle missing  values
df=df.fillna(df.mean())
```

```python
df.isnull().sum() # here  we   again  check  for  any   missing   value  and  we  see  no  missing    value  find
```

```python
# separate dataset  into  independent  and   dependent variables
X=df.drop("student_marks",axis=1) #  here   we   remove   student_marks  from  the  table  , contains  independent  variabke

y=df.drop("study_hours",axis=1) # contains  dependent variable
```

```python
X.head()
```

```python
y.head()
```

```python
# now  we  split  the  dataset  into  training  and   testing   dataset
```

```python
#Training test  spliting
#here  x  divides  into  parts and  y   divides  into  2   parts
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
#  now  here  we  apply  linear   regression algorithm - (y=mx +c)
from   sklearn.linear_model import  LinearRegression 
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
#testing the  model
sh=float(input("enter  study  hours  from 3-12 : ")) # sh -  study  hours 
if sh>=3 and sh<=12 :
    res=lr.predict([[sh]])[0][0].round(2)
    print("Expected marks]=",res)
else:
    print("invalid input")
```
